### 주식시장 주요지표 결과에 따른 예측


```python
def stock_market(pmi, cpi, forecast_cpi):
    # 1. PMI 판단 (기준 50)
    if pmi >= 50:
        print(f"PMI {pmi}: 경기 확장 국면입니다. (기업 활동 활발)")
    else:
        print(f"PMI {pmi}: 경기 위축 국면입니다. (주의 필요)")

    # 2. CPI 판단 (예상치와 비교)
    # 오타 수정: forcast_cpi -> forecast_cpi
    if cpi > forecast_cpi:
        print(f"CPI {cpi}%: 시장 예상({forecast_cpi}%) 상회! 물가 상승 및 금리 인상 우려가 있습니다.")
    elif cpi < forecast_cpi:
        print(f"CPI {cpi}%: 시장 예상({forecast_cpi}%) 하회! 물가 안정 및 금리 인하 기대감이 있습니다.")
    else:
        print(f"CPI {cpi}%: 예상치와 부합합니다. 시장이 안도할 가능성이 큽니다.")

# 실행 결과: PMI 52.3(확장), CPI 2.4(예상 2.5보다 낮음 -> 호재)
stock_market(52.3, 2.4, 2.5)
```
